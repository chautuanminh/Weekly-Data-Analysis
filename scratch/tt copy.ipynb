```python
import pandas as pd
import requests
import numpy as np
from bs4 import BeautifulSoup
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from webdriver_manager.chrome import ChromeDriverManager
import time
import shutil
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.common.by import By
```

```python
chrome_path = shutil.which("chromium-browser") or shutil.which("chromium")
driver_path = shutil.which("chromedriver")
```

```python
# Configure Chrome options
options = Options()
options.add_argument("--start-maximized")   # open browser maximized
# options.add_argument("--headless=new")    # run without opening a window (optional)

# Start WebDriver
service = Service(ChromeDriverManager().install())
driver = webdriver.Chrome(service=service, options=options)

# Test by opening a webpage
driver.get("https://batdongsan.com.vn/nha-dat-ban")
print("Page title:", driver.title)
```

```python

listings = driver.find_elements(By.CSS_SELECTOR, "div.js__card")

data = []

for listing in listings:
    try:
        # link
        url = listing.find_element(By.CSS_SELECTOR, "a.js__product-link-for-product-id").get_attribute("href")
    except:
        url = None

    try:
        # title
        title = listing.find_element(By.CLASS_NAME, "re__card-info").get_attribute("title")
    except:
        title = None

    try:
        # image (first image)
        image = listing.find_element(By.CSS_SELECTOR, "div.re__card-image img").get_attribute("src")
    except:
        image = None

    try:
        price = listing.find_element(By.CSS_SELECTOR, "span.re__card-config-price.js__card-config-item").text
    except:
        price = None

    try:
        area = listing.find_element(By.CSS_SELECTOR, "span.re__card-config-area.js__card-config-item").text
    except:
        area = None

    try:
        price_per_m2 = listing.find_element(By.CSS_SELECTOR, "span.re__card-config-price_per_m2.js__card-config-item").text
    except:
        price_per_m2 = None

    try:
        bedrooms = listing.find_element(By.CSS_SELECTOR, "span.re__card-config-bedroom.js__card-config-item span").text
    except:
        bedrooms = None

    try:
        toilets = listing.find_element(By.CSS_SELECTOR, "span.re__card-config-toilet.js__card-config-item span").text
    except:
        toilets = None

    try:
        location = listing.find_element(By.CSS_SELECTOR, "div.re__card-location span:nth-of-type(2)").text
    except:
        location = None

    # --- Save all into dictionary ---
    data.append({
        "title": title,
        "url": url,
        "image": image,
        "price": price,
        "area": area,
        "price_per_m2": price_per_m2,
        "bedrooms": bedrooms,
        "toilets": toilets,
        "location": location
    })

# --- Print results ---
for d in data:
    print(d)

driver.quit()
```

```python
df = pd.DataFrame(data)

# Save to CSV
df.to_csv("bdsbds.csv", index=False, encoding="utf-8-sig")

print(df.head())
```

```python
cards = driver.find_elements(
    "css selector", "div.js__project-card.js__card-project-web.re__prj-card-full"
)

data = []


for card in cards:
    try:
        # Title
        title = card.find_element(By.CLASS_NAME, "re__prj-card-title").text

        # Location
        location = card.find_element(By.CLASS_NAME, "re__prj-card-location").text

        # Area (first config value)
        area = card.find_elements(By.CLASS_NAME, "re__prj-card-config-value")[0].text

        # Company
        company = card.find_element(By.CLASS_NAME, "re__span-full-max-width").text

        print("Title:", title)
        print("Location:", location)
        print("Area:", area)
        print("Company:", company)
        print("-" * 50)
    except Exception as e:
        print("Error extracting a card:", e)
    data.append({
        "Title": title,
        "Location": location,
        "Area": area,
        "Company" : company,
    })

# Convert to DataFrame
```

```python
df = pd.DataFrame(data)

# Save to CSV
df.to_csv("batdongsan_projects.csv", index=False, encoding="utf-8-sig")

print(df.head())
```
